#### Custom Google Search Function

```python
from requests import get
from bs4 import BeautifulSoup


def search(term, num_results=10, lang="nl"):
    usr_agent = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:40.0) Gecko/20100101 Firefox/40.1'}

    def fetch_results(search_term, number_results, language_code):
        escaped_search_term = search_term.replace(' ', '+')

        google_url = 'https://www.google.com/search?q={}&num={}&hl={}'.format(escaped_search_term, number_results+1,language_code)
#         proxies = {
#             'http': 'socks5://Username:Password@IP1:20000',
#             'https': 'socks5://Username:Password@IP1:20000'
#         }
        response = get(google_url, headers=usr_agent) #, proxies=proxies
        response.raise_for_status()
        if response.status_code == 429:
            print(response.headers["Retry-After"])

        return response.text

    def parse_results(raw_html):
        soup = BeautifulSoup(raw_html, 'html.parser')
        result_block = soup.find_all('div', attrs={'class': 'g'})
        for result in result_block:
            link = result.find('a', href=True)
            title = result.find('h3')
            if link and title:
                yield link['href']

    html = fetch_results(term, num_results, lang)
    return list(parse_results(html))
```

```python
from googlesearch import search
# Getting urls, storing them as a list of lists 
#Use apply to throw a function over each keyword https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html
urls = []
for kw in df['Nederland', 'Keyword ']:
    print("Current KW:  ", kw)
    temp = []
    for j in search(kw, lang='nl'):
        temp.append(j)
        while len(temp) > 10:
            temp.pop()
    urls.append(temp)
```

```python
from urllib.parse import urlparse

domains = []
for urllists in urls:
    temp=[]
    for url in urllists:
        parsed = urlparse(url)
        domain = '{uri.scheme}://{uri.netloc}/'.format(uri=parsed)
        temp.append(domain)
    domains.append(temp)
# print(domains)
# print(len(domains))

#Get tld's
tlds = []
for domainlist in domains:
    temp = []
    for domain in domainlist:
        tld = domain.rpartition('.')[2]
        temp.append(tld)
    tlds.append(temp)


```

```python
# Append new lists
# urls = pd.Series(list(range(21)))
df['Nederland', 'urls'] = urls
df['Nederland', 'domains'] = domains
df['Nederland', 'tlds'] = tlds
df = df.sort_index(axis=1)
df.head(3)
```
